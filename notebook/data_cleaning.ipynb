### Data Quality Dimensions

- Completeness -> is data missing?
- Validity -> is data invalid -> negative height -> duplicate patient id
- Accuracy -> data is valid but not accurate -> weight -> 1kg
- Consistency -> both valid and accurate but written differently -> New Youk and NY

### Data Cleaning Order

1. Quality -> Completeness
2. Tidiness
3. Quality -> Validity
4. Quality -> Accuracy
5. Quality -> Consistency

#### Steps involved in Data cleaning
- Define
- Code
- Test

```python
df.loc[df["os"].isnull(),"os"]="Android 11"#as we see in the name
```

```python
for i in df["os"]:
  if "Windows" in i:
    df["os"].replace(i,"Windows",inplace=True)
  elif "Mac" in i:
    df["os"].replace(i,"Mac",inplace=True)
  else:
    df["os"].replace(i,"Others",inplace=True)
```

```python
for i in df["aspect_ratio"]:
  if i!="16:10" and i!="3:2":
    df["aspect_ratio"].replace(i,"16:9",inplace=True)
```

```python
for i in df["core"]:
  if "Core" in i:
    pass
  else:
    df["core"].replace(i,np.nan,inplace=True)
```

```python
col_list=["antiglare","touch_screen","hdmi","mcr","bluetooth","backlit_keyboard","inbuilt_microphone","thunderbolt","fingerprint_sensor","ethernet","display_port"]
for i in col_list:
  df[i] = df[i].apply(lambda x: 0 if pd.isnull(x) else 1)
```

```python
df["wifi"]=df["wifi"].apply(lambda x:0 if x!="WiFi" else 1)
```

```python
df["hdd"]=df["hdd"].apply(lambda x:"0" if pd.isnull(x) else x)
for i in df["hdd"]:
  if "TB" in i:
    df["hdd"].replace(i,i.split("TB")[0],inplace=True)
  elif "GB" in i:
    df["hdd"].replace(i,i.split("GB")[0],inplace=True)
  else:
    df["hdd"].replace(i,np.nan)
```

```python
df["hdd"]=df["hdd"].astype(int)
for i in df["hdd"]:
  if i<=10:
    df["hdd"].replace(i,i*1024,inplace=True)
```

```python
df["ssd"]=df["ssd"].apply(lambda x:"0" if pd.isnull(x) else x)
for i in df["ssd"]:
  if "TB" in i:
    df["ssd"].replace(i,i.split("TB")[0],inplace=True)
  elif "GB" in i:
    df["ssd"].replace(i,i.split("GB")[0],inplace=True)
```

```python
df["ssd"]=df["ssd"].astype(int)
for i in df["ssd"]:
  if i<=10:
    df["ssd"].replace(i,i*1024,inplace=True)
```

```python
df["graphic"].fillna("Integrated",inplace=True)
for i in df["graphic"]:
  if "NVIDIA" in i:
    df["graphic"].replace(i,"NVIDIA",inplace=True)
  elif "AMD" in i:
    df["graphic"].replace(i,"AMD",inplace=True)
  else:
    df["graphic"].replace(i,"Integrated",inplace=True)
```

```python
df["camera"]=df["camera"].str.replace("MP","").str.strip()
```

```python
df["usb"].fillna("0x0",inplace=True)
```

```python
def extract_ports(port_string):
    usb3 = usb2 = typec = 0
    parts = port_string.split(' x ')
    if len(parts)==2:
      if 'USB 3.0' in parts[1]:
          usb3 = int(parts[0])
      elif 'USB 2.0' in parts[1]:
          usb2 = int(parts[0])
      elif 'Type-C' in parts[1]:
          typec = int(parts[0])
    return pd.Series([usb3, usb2, typec])
```

```python
df[["usb3", "usb2", "type_c"]] = df["usb"].apply(extract_ports)
```

```python
df["core"].fillna("",inplace=True)
```

```python
def convert_cores(core_string):
    if "Cores" in core_string:
        return int(core_string.split(" Cores")[0])
    elif "Quad" in core_string:
        return 4
    elif "Octa" in core_string:
        return 8
    elif "Hexa" in core_string:
        return 6
    elif "Dual" in core_string:
        return 2
    else:
        return np.nan
```

```python
df["core"]=df["core"].apply(convert_cores)
```

```python
df["processor"].value_counts()
```

```python
def extract_processor_brand(processor):
  processor=processor.lower()
  if "intel" in processor:
    return "Intel"
  elif "amd" in processor:
    return "AMD"
  elif "apple" in processor:
    return "Apple"
  elif "mediatek" in processor:
    return "Mediatek"
```

```python
df["processor_brand"]=df["processor"].apply(extract_processor_brand)
```

```python
def extract_processor_model(processor):
  processor=processor.split()
  if len(processor)>=6:
    return processor[4]
  else :
    return processor[-1]

```

```python
df["processor_model"]=df["processor"].apply(extract_processor_model)
```

```python
df["processor_model"].replace("Chip","M2",inplace=True)
df["processor_model"].replace("Pro","M2",inplace=True)
df["processor_model"].replace("Max","M2",inplace=True)
df["processor_model"].replace("Silver","3050U",inplace=True)
```

```python
def extract_processor_generation(processor):
  processor=processor.split()
  return processor[0]
```

```python
df["processor_gen"]=df["processor"].apply(extract_processor_generation)
```

```python
df.drop(columns=["processor","usb"],inplace=True)
```

```python
val_list=["intel","mediatek","amd","apple"]
df[df["processor_gen"].str.lower().isin(val_list)]
```
